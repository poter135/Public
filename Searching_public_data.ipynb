**使用python爬蟲爬回以下網址 (https://mops.twse.com.tw/mops/web/t05sr01_1)**
- 主旨含董事會及股利的資料，並組成一個表格
- 表格欄位含(代號、公司簡稱、發言日期、主旨、詳細資料)


```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import re
```

```python
url = 'https://mops.twse.com.tw/mops/web/t05sr01_1'
headers = {'user-agent': 'Mozilla/5.0'}
r= requests.get(url,headers=headers).text
soup = BeautifulSoup(r,'html.parser')

even = soup.find_all(class_ = 'even')
odd = soup.find_all(class_ = 'odd')

code = []
company = []
date = []
sub = []
for i in range(len(even)+len(odd)):
    if(i%2==0):
        info = even[int(i/2)].text
        ls = re.findall("\n.+",info)
        code.append(ls[0][1:])
        company.append(ls[1][1:])
        date.append(ls[2][1:])
        sub.append(ls[4][1:])
    else:
        info = odd[int(i/2)].text
        ls = re.findall("\n.+",info)
        code.append(ls[0][1:])
        company.append(ls[1][1:])
        date.append(ls[2][1:])
        sub.append(ls[4][1:])
        
df = pd.DataFrame({"代號":code,"公司簡稱":company,"發言日期":date,"主旨":sub})
df
```

```python
df.to_csv('data.csv',index = False)
```
